**인공지능 실습 1_2 - 뉴런 구현(Perceptron using Tensorflow Keras API)**

```python
import numpy as np
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='sigmoid') # 뉴런 하나 정의
])
```

```python
model.compile(optimizer='SGD', loss='MeanSquaredError')
```

```python
# training set
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])
```

```python
model.fit(X, y, epochs=10000, verbose=0) # verbose : training 중간 결과를 어느정도 볼지
```

```python
z = model.predict(X)
print(z) # 0 0 0 1 (확률 높은 것이 1)
```

```python
model.summary() # 만들어진 모델의 architecture 보여줌 
```
