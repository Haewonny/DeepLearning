**인공지능 실습 1_3 - MLP 구현(MLP using Tensorflow Keras API)**

```python
import numpy as np
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Dense(3, activation='tanh'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='SGD', loss=tf.keras.losses.BinaryCrossentropy())
```

```python
# training set
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])
```

```python
model.fit(X, y, epochs=10000, verbose=0) # verbose : training 중간 결과를 어느정도 볼지
```

```python
z = model.predict(X)
print(z) # 0 1 1 0
```
